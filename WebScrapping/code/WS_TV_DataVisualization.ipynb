### 5. Calculating Mean values of the 'Discounted Price' to Compare over each 'Brand Name' & 'Screen Size' using Barplot

```python
# Calculating Mean values of the 'Discounted Price' to Compare over each 'Brand Name' using Barplot

def compare_means(df,discrete_col,continuous_col):
    group = df.groupby([discrete_col],as_index=False)[continuous_col].mean().reset_index(drop=True)
    plt.figure(figsize=(20,8))
    sns.barplot(x=group[discrete_col],y=group[continuous_col],palette='twilight_r')
    plt.ylabel('mean ' + continuous_col)
    plt.show()

compare_means(df,'Brand Name','Discounted Price')
```

```python
# Calculating Mean values of the 'Discounted Price' to Compare over each 'Screen Size' using Barplot

def compare_means(df,discrete_col,continuous_col):
    group = df.groupby([discrete_col],as_index=False)[continuous_col].mean().reset_index(drop=True)
    plt.figure(figsize=(20,8))
    sns.barplot(x=group[discrete_col],y=group[continuous_col],palette='twilight_shifted')
    plt.ylabel('mean ' + continuous_col)
    plt.show()

compare_means(df,'Screen Size','Discounted Price')
```

## Bivariate Analysis Using Multivariate Dataset plotting Heatmap & Pairplot


### 1. Heatmap Using Groupby and Pivot of 'Brand Name & Screen Size' over 'Discount Percentage'

```python
def feature_interactions(df,feature1, feature2,continuous_col):
    group = df.groupby([feature1,feature2],as_index=False)[continuous_col].mean().reset_index(drop=True)
    pivot = group.pivot(index=feature1, columns=feature2, values=continuous_col)
    pivot.fillna(0, inplace=True)
    plt.figure(figsize=(20,8))
    sns.heatmap(pivot,cmap='gist_rainbow_r')
    plt.show()

feature_interactions(df,'Brand Name','Screen Size','Discount Percentage')
```
