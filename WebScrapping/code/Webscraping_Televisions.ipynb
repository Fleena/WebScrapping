# WebScraping - Televisions

### In this notebook, we will webscrape the data from flipkart and export the data for further data cleaning and analysis.

```python
import bs4
from bs4 import BeautifulSoup as bs
import requests
import time
import pandas as pd
import missingno as mino
```

```python
urls = ["https://www.flipkart.com/search?q=latest+4k+smart+tv&otracker=AS_Query_HistoryAutoSuggest_2_0&otracker1=AS_Query_HistoryAutoSuggest_2_0&marketplace=FLIPKART&as-show=on&as=off&as-pos=2&as-type=HISTORY&p%5B%5D=facets.rating%255B%255D%3D4%25E2%2598%2585%2B%2526%2Babove&p%5B%5D=facets.brand%255B%255D%3Drealme&p%5B%5D=facets.brand%255B%255D%3DOnePlus&p%5B%5D=facets.brand%255B%255D%3DThomson&p%5B%5D=facets.brand%255B%255D%3DMi&p%5B%5D=facets.brand%255B%255D%3DCoocaa&p%5B%5D=facets.brand%255B%255D%3DCroma&p%5B%5D=facets.brand%255B%255D%3DHaier&p%5B%5D=facets.brand%255B%255D%3DiFFALCON%2Bby%2BTCL&p%5B%5D=facets.brand%255B%255D%3DNokia&p%5B%5D=facets.brand%255B%255D%3DMOTOROLA&p%5B%5D=facets.brand%255B%255D%3DMicromax&p%5B%5D=facets.brand%255B%255D%3DLloyd&p%5B%5D=facets.brand%255B%255D%3DONIDA&p%5B%5D=facets.brand%255B%255D%3DOxygen&p%5B%5D=facets.brand%255B%255D%3DPHILIPS&p%5B%5D=facets.brand%255B%255D%3DSAMSUNG&p%5B%5D=facets.brand%255B%255D%3DTOSHIBA&p%5B%5D=facets.brand%255B%255D%3DVu&p%5B%5D=facets.discount_range_v1%255B%255D%3D10%2525%2Bor%2Bmore&page="+str(i) for i in range(1,9)]
headers = ({'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.79 Safari/537.36 Edge/14.14393'})
```

#### Extracting all the configurations

```python
# Initialization of the lists to store the extracted data
# The data that we extract is unstructured data. So we’ll create empty lists to store them in a structured form
count=0                  
products=[]              
prices=[]                
ratings=[]               
applications = []                 
opersys = []                 
resolution = []                  
speakerop = []                  
refreshr = []
discount = []
```

```python
for link in urls:
    page = requests.get(link , headers = headers)
    print(link, 'status:', page)
    soup = bs(page.content, 'html.parser')        
   
    
    for info in soup.findAll('a',class_='_1fQZEK'):
        name = info.find('div', attrs={'class':'_4rR01T'})
        disc = info.find('div', attrs={'class':'_30jeq3 _1_WHN1'})
        rating = info.find('div', attrs={'class':'_3LWZlK'})
        price = info.find('div', attrs={'class':'_3I9_wc _27UcVY'})
        
        ## Splitting integrated specification into individual specifications
        spec = info.find('div', attrs={'class':'fMghEO'})    
        for col in spec:
            col = col.find_all('li', attrs={'class':'rgWa7D'})
            apps = col[0].text
            os = col[1].text
            pixels = col[2].text
            speaker = col[3].text
            rr = col[4].text
                        
        
        products.append(name.text)
        prices.append(price)
        
        applications.append(apps)
        opersys.append(os)
        resolution.append(pixels)
        speakerop.append(speaker)
        refreshr.append(rr)
        discount.append(disc)
       
        
        ratings.append(rating.text) if type(rating) == bs4.element.Tag  else ratings.append('NaN')
        
        count=count+1

    di = {'Product Name':products,'Supported Apps':applications,'OS':opersys,'Resolution':resolution,'Speaker Output':speakerop,'Refresh Rate':refreshr,'Rating':ratings,'Original Price':prices,'Discounted Price':discount}
    tvlist_df = pd.DataFrame.from_dict(di, orient='index')
    
    time.sleep(20)
```

#### Applying Transpose to append the datas in structured format

```python
tvlist_T = tvlist_df.T
```
